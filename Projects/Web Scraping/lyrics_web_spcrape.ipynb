```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

```python
song_list = "https://www.lyrics.com/artist/Kanye-West/353484"
data = requests.get(song_list)
soup = BeautifulSoup(data.text)
```

```python
tables = soup.select('table.tdata')
links = []
for t in tables:
    links_page = t.find_all('a')
    links_page = [l.get("href") for l in links_page]
    links_page = [l for l in links_page if '/lyric/' in l] 
    for l in links_page:
        links.append(l)

links = [*set(links)]
print(links)
```

```python
song_div = soup.select('div.col-xs-12.col-lg-8.text-center')[0]
links = song_div.find_all('a')  # all links in the web page
links = [l.get("href") for l in links]  # get the reference to all the links
links = [l for l in links if isinstance(l, str)]
links = [l for l in links if '/lyrics/' in l]
song_urls = [f"https://azlyrics.com{l}" for l in links] # all absolute links to songs
```

```python
song_url = song_urls[0] # first club page link
data = requests.get(song_url)
soup = BeautifulSoup(data.text) # extract all information from page
```
